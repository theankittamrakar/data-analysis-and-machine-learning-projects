We can see that the dataset has values in strings.We need to convert all the unique values to integers. Thus we perform label encoding on the data.

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
 
data.tail(10)
```

```python
##for col in data.columns:
    #print (data[col].unique())
for col in data.columns:
    print('\n*', col, '*')
    print(data[col].value_counts())
```

The 'veil-type' column has only one value, '0', which means, every mushroom has the same veil-type. because of that, the column doesnt have any useful information - so we can drop that column.

```python
data.drop(['veil-type'], axis=1, inplace=True)
```

```python
## Seperating into labels and features
X = data.iloc[:,1:23]  # all rows, all the features and no labels
y = data.iloc[:, 0]  # all rows, label only
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
from sklearn import metrics
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(metrics.classification_report(y_test,y_pred))
```
