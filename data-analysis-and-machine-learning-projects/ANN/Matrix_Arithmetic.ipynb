# Matrix Multiplication:

To define the matrix product of matrices  A and B, A  must have the same number of columns as  B . If  A  is of shape m x n and  B  is of shape n x p, then  C  is of shape m x p.

mat_mul.jpg


```python
# Create a matrix A

A = np.array([[0, 1, 1], [1, 0, 1]])
A
```

```python
# Create a matrix B

B = np.array([[1, 1], [1, 1], [-1, 1]])
B
```

```python
assert A.shape[1] == B.shape[0], "Matrix multiplication is Not possible"
```

```python
#Perform Matrix multiplication
#solution

Z = np.dot(A,B)
print(Z.shape)
```
